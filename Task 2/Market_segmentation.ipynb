```python
dp=pd.read_csv("/content/mcdonalds.csv")
```

```python
print("Duplicated value on the dataset : ",dp.duplicated().sum())
```

```python
dp.drop_duplicates(keep="first",inplace=True)
```

```python
dp
```

```python
dp["Gender"]=dp["Gender"].replace({"Male":1,"Female":0})
dp["yummy"]=dp["yummy"].replace({"Yes":1,"No":0})
dp["convenient"]=dp["convenient"].replace({"Yes":1,"No":0})
dp["spicy"]=dp["spicy"].replace({"Yes":1,"No":0})
dp["fattening"]=dp["fattening"].replace({"Yes":1,"No":0})
dp["greasy"]=dp["greasy"].replace({"Yes":1,"No":0})
dp["fast"]=dp["fast"].replace({"Yes":1,"No":0})
dp["cheap"]=dp["cheap"].replace({"Yes":1,"No":0})
dp["expensive"]=dp["expensive"].replace({"Yes":1,"No":0})
dp["healthy"]=dp["healthy"].replace({"Yes":1,"No":0})
dp["tasty"]=dp["tasty"].replace({"Yes":1,"No":0})
dp["disgusting"]=dp["disgusting"].replace({"Yes":1,"No":0})
dp["VisitFrequency"]=dp["VisitFrequency"].replace({"Every three month":1,"Once":0})
dp["tasty"]=dp["tasty"].replace({"Yes":1,"No":0})


dp
```

```python
plt.figure(figsize=(12,9))
s=sns.heatmap(dp.corr(),annot=True)
```

```python
plt.figure(1,figsize=(15,7))
sns.set(style="darkgrid")
ax=sns.countplot(x="Gender",data=dp,hue="VisitFrequency")
```

```python
plt.figure(1,figsize=(12,7))
sns.set(style="darkgrid")
ax=sns.countplot(x="tasty",data=dp,hue="VisitFrequency")
```
